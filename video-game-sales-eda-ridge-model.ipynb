This dataset contains a list of video games with sales greater than 100,000 copies. It was generated by a scrape of vgchartz.com.


**Fields include**

* Rank - Ranking of overall sales
* Name - The games name
* Platform - Platform of the games release (i.e. PC,PS4, etc.)
* Year - Year of the game's release
* Genre - Genre of the game
* Publisher - Publisher of the game
* NA_Sales - Sales in North America (in millions)
* EU_Sales - Sales in Europe (in millions)
* JP_Sales - Sales in Japan (in millions)
* Other_Sales - Sales in the rest of the world (in millions)
* Global_Sales - Total worldwide sales.

There are 16,598 records. 2 records were dropped due to incomplete information.

# Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
df.Year=df.Year.astype(object)
df.Rank=df.Rank.astype(object)
for i in ['Name','Platform','Year','Genre','Publisher']:
    model=encoder.fit(df[i])
    df[i]=model.transform(df[i])
df.head()
```

# Split Train Data and Test Data

```python
from sklearn.model_selection import train_test_split

train, test=train_test_split(df,test_size=0.3,shuffle=True,random_state=50)
print('Shape of Train Data:',train.shape,'\nShape of Test Data:',test.shape)
```

# Normalizing

```python
for i in ['Global_Sales','NA_Sales','EU_Sales','JP_Sales','Other_Sales']:
    train[i]=(train[i]-train[i].mean())/train[i].std()
    test[i]=(test[i]-train[i].mean())/train[i].std()
train[['Global_Sales','NA_Sales','EU_Sales','JP_Sales','Other_Sales']].head()
```

# Data Preparation

```python
train_x=train.drop(columns='Global_Sales')
train_y=train.Global_Sales
test_x=test.drop(columns='Global_Sales')
test_y=test.Global_Sales
print(train_x.shape,train_y.shape,'\n',test_x.shape,test_y.shape)
```

# Modeling

```python
from sklearn.linear_model import RidgeCV
from sklearn.model_selection import KFold,cross_val_score
from sklearn.preprocessing import RobustScaler
from sklearn.pipeline import make_pipeline

kf=KFold(n_splits=10,random_state=50,shuffle=True)

def rmse(model,x,y):
    rmse=np.sqrt(-cross_val_score(model,x,y,scoring='neg_mean_squared_error',cv=kf).mean())
    return rmse

ridge = make_pipeline(RobustScaler(), RidgeCV(cv=kf))
ridge=ridge.fit(train_x,train_y)

train_score=rmse(ridge,train_x,train_y)
test_score=rmse(ridge,test_x,test_y)

print('Train Score =',train_score,"\nTest Score =",test_score)
```
