```python
import pandas as pd
import numpy as np
import plotly.express as px
```

```python
df = pd.read_csv('vgsales.csv')
```

```python
fig = px.scatter(df, x='Global_Sales', y='EU_Sales', animation_frame='Year', size='JP_Sales', color='Genre', hover_name='Name', log_x=True)

fig.show()
```

```python
# Create a pie chart of the genre column
fig = px.pie(df, names="Genre")

# Show the chart
fig.show()

# title='Global Sales by Platform')
```
